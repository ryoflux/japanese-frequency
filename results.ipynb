```python
import matplotlib.pyplot as plt
import json
import statistics
```

```python
with open("combined_results.json") as f:
  data = json.load(f)
```

```python
number = []
count = []
for i,(_, v) in enumerate(data['characters'].items()):
  if v['script'] == 'kanji':
    number.append(i)
    count.append(v['count'])

print(len(count))
```

```python
plt.figure(dpi=1200)
plt.plot(number, count)
plt.title('Distribution of kanji in three texts')
plt.ylabel('Number of times used')
plt.xlabel('Amount of kanji')
plt.ylim([0,7313])
plt.xlim([0, 2412])
plt.show()
```

```python
counts = [v['count'] for _, v in data['characters'].items() if v['script'] == 'kanji']

print(f"Mean number of times kanji used: {statistics.mean(counts)}")
print(f"Mode number of times kanji used: {statistics.mode(counts)}")
print(f"Median number of time kanji used: {statistics.median(counts)}")
print(f"Number of kanji used once {len([i for i in counts if i == 1])}")
```
